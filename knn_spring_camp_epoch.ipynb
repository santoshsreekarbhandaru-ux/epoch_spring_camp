```python
import numpy as np
```

```python
data = [
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
]
data
```

```python
# encoding 
# 'Apple' → 0
# 'Banana' → 1
# 'Orange' → 2
labels=[]
for row in data:
    i=row[3]
    labels.append(i)
  
def encode(col):
    labels1=[]
    for i in col:
        if i =='Apple':
            labels1.append(0)
        if i =='Banana':
            labels1.append(1)
        if i =='Orange':
            labels1.append(2)
    return labels1
print(encode(labels))
# we can use one-hot coding here :) in place of this 
```

```python
def decode(col):
    labels1=[]
    for i in col:
        if i ==0:
            labels1.append('Apple')
        if i ==1:
            labels1.append('Banana')
        if i ==2:
            labels1.append('Orange')
    return labels1
```

```python
X_train=[row[:3] for row in data]
Y_train=encoding(labels)
X_train,Y_train
```

```python
##euclidean distance
def euclid_dist(p1,p2):
    dist=np.sqrt(np.sum((np.array(p1)-np.array(p2))**2))
    return dist

## manhattan 
def manhattan_dist(p1,p2):
    dist=(np.sum(abs(np.array(p1)-np.array(p2))))
    return dist
```

```python
#example to test the function
a=[2,4]
b=[2,6]
print(euclid_dist(a,b))
print(manhattan_dist(a,b))
```

```python
class KNN:
    def __init__(self,k=3):#stores value of k
        self.k=k
        
    def fit(self,X,y):#stores training data
        self.X_train=np.array(X)
        self.y_train=np.array(y)
        
    def predict(self,X_test):#Returns predictions for an array of test points
        d=np.array(X_test)
        predictions=[self.predict_one(x) for x in d]
        return predictions
        
    def predict_one(self,x): #Predicts the label for a single test sample
        x=np.array(x)
        dist=[] #it's an array of the distances and the labels from every point to test sample
        for i in range(len(self.X_train)):
            q=euclid_dist(x,self.X_train[i])
            dist.append((q,self.y_train[i]))
            
        dist.sort(key=lambda x:x[0]) #sorting array to find first k least distances
        
        k_nearest=dist[:self.k] #first k distances and their labels
        
        #extracting k nearest labels from k_nearest
        k_labels=[]
        for l in k_nearest:
            k_labels.append(l[1])
            
        #fing maximum repeated label in k_labels
        max_count=0
        max_label=None
        counts={}
        #we can also use Counter function for this
        for label in k_labels:
            if label not in counts:
                counts[label]=0
            counts[label]+=1
        for label in counts:
            if counts[label]>max_count:
                max_count=counts[label]
                max_label=label
                
        return max_label
        
        
    
```

```python
model=KNN(3)
model.fit(X_train,Y_train)
```

```python
test_data = np.array([
    [118, 6.2, 0],  # Expected: Banana
    [160, 7.3, 1],  # Expected: Apple
    [185, 7.7, 2]   # Expected: Orange
])

X_test=[row[:3] for row in test_data]
y_test=['Banana','Apple','Orange']
```

```python
y_pred=model.predict(X_test)
y_preds=decode(y_pred)
y_preds
```

```python
##bonus 
##1 accuracy checker
def accuracy_check(y_pred,y_test):
    ans=0
    for i in range(len(y_pred)):
        if y_pred[i]==y_test[i]:
            ans+=1
    return (ans/len(y_pred))*100
accuracy_check(y_preds,y_test)
```
